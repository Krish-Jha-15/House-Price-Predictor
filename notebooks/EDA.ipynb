## Exploratory Data Analysis(EDA)


#Step which are been done
1.   Using df.head
2.   Ploting the distribution of feature
3.   Find Correlation and plotting
4.   Histplot of target
5.   scattered distribution of features and target



```python
import pandas as pd
import matplotlib
%matplotlib inline

df=pd.read_csv('../data/Housing_Data-set.csv')

df.head()
```

```python
df.describe()
```

#Ploting the distribution of feature

```python
import matplotlib.pyplot as plt

df.hist(figsize=(16,10))
plt.show()
```

#Correlation of each feature

```python
import seaborn as sns

corr = df.corr(numeric_only=True)
corr

sns.heatmap(corr,annot=False,cmap='coolwarm')
plt.figure(figsize=(12,8))
plt
```

#Distribution of Target

```python
df['price'].describe()
sns.histplot(df['price'],kde=True)
plt.show()
```

```python
sns.scatterplot(x='sqft_living',y='price',data=df)
plt.show()
```

```python
sns.scatterplot(x='bedrooms',y='price',data=df)
plt.show()
```

```python
sns.boxplot(x='waterfront',y='price',data=df)
plt.show()
```
