##Outliers

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
df.boxplot()
plt.show()
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['price'] < lower) | (df['price'] > upper)]

print("Outliers:", len(outliers))
```

ther are **240** Outliers But we will not remove them from data set as the values are realistic houses
